Data from https://www.kaggle.com/datasets/arslanali4343/covid19-data-from-world

```python
from IPython.display import display
import pandas as pd
import numpy as np
from natsort import index_natsorted
df = pd.read_csv("owid-covid-data.csv")
df
```

```python
location_name = df.drop_duplicates(subset=['location']) 

location	= []
sum_case = []
continent = []
no_need = ['Upper middle income','Lower middle income','High income','European Union']

for l in location_name['continent']:
  continent.append(l)

for i in location_name['location']:
  if i not in continent and i != 'World' and i not in no_need:
    location.append(i)

for j in location:
  case_sum = df[df['location'] == j]
  sum_case.append(case_sum['new_cases'].sum())
```

```python
import matplotlib.pyplot as plt
x = location
y = sum_case

plt.figure(figsize=(10,50))
plt.barh(x,y, color= "red")
plt.grid()
plt.show()
```
