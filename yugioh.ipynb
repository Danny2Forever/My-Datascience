```python
from IPython.display import display
import pandas as pd
import numpy as np
from natsort import index_natsorted
df = pd.read_csv("Yu-Gi-Oh-Normal_Monsters.csv")
```

```python
attack_lst = []
defense_lst = []
for i in df['Attack_Points']:
  try :                                      #Delete non-number value
    int(i) 
    attack_lst.append(int(i))
  except:
    df = df[df.Attack_Points != i]

for j in df['Defense_Points']:
  try :                                      #Delete non-number value 
    int(j) 
    defense_lst.append(int(j))
  except:
    df = df[df.Defense_Points != j]

df = df[df.Name != 'Obelisk the Tormentor (Non-Playable)']  #Obelisk becuase it's not normal monster

print(f"There are have {len(df.index)} mosters from Dataset.")
print(f"The the max Attack Points is {np.max(attack_lst)}")
print(f"The the max Defense Points is {np.max(defense_lst)}")
```

<h2>⚔️ Sorted by Attack Points.</h2>



```python
df = df.sort_values(
      by="Attack_Points",
      ascending=False,
      key=lambda x: np.argsort(index_natsorted(df["Attack_Points"])))
df
```

<h2>🛡️ Sorted by Defense Points.</h2>


```python
df = df.sort_values(
      by="Defense_Points",
      ascending=False,
      key=lambda x: np.argsort(index_natsorted(df["Defense_Points"])))
df
```

<h2>⭐ Sorted by Level.</h2>

```python
df = df.sort_values(by="Level", ascending=False)
df
```
